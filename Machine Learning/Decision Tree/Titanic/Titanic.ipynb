```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
from sklearn.tree import plot_tree
from sklearn.metrics import mean_absolute_error, mean_squared_error

#os.chdir('C:\\Users\Matthew...')
path = os.getcwd()

df = pd.read_csv('titanic.csv')
```

```python
#Show the columns with null values (yellow)
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#Drop the values with high null values no impact
df = df.drop("boat", axis=1)  
df = df.drop("body", axis=1)
df = df.drop('cabin', axis=1)
df = df.drop('home.dest', axis=1)
```

```python
#Function to process and store an estimated age
def impute_age(cols):   
    Age = cols[0]    
    Pclass = cols[1]       
    if pd.isnull(Age):
        if Pclass == 1:           
            return 37
        elif Pclass == 2:           
            return 29
        else:           
            return 24
    else:        return Age
```

```python
#Apply function to the age column
df['age'] = df.apply(impute_age, axis=1)
```

```python
#Now nearly all non-null values
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#Drop the non applicable to get all non-null counts shown with info
df = df.dropna()
df.info()
```

```python
#Train Test Split
X = pd.get_dummies(df.drop("survived",axis=1),drop_first=True)
y = df["survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#Fit model to decision tree classifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
base_pred = model.predict(X_test)
```

```python
#Create and plot the confusion matrix
confusion_matrix(y_test,base_pred)
```

```python
#Plot the normalized confusion matrix
plot_confusion_matrix(model,X_test,y_test, normalize="true")
```

```python
#Display a pruned tree (entire tree is massive)
pruned_tree = DecisionTreeClassifier(max_depth = 3)
pruned_tree.fit(X_train,y_train)

plt.figure(figsize=(12,8),dpi=150)
plot_tree(pruned_tree,filled=True,feature_names=X.columns);
```

```python
#Show classification report data
print(classification_report(y_test,base_pred))
```
