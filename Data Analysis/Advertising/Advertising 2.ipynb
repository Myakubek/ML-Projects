```python
#Advertising Example
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

#os.chdir('C:\\Users\Matthew...')
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head(10)
```

```python
sns.pairplot(df, corner=True)#without corner = true product all variations
#Can see from below the more spent on TV the more sales there are
#Radio seems less correlated, newspaper even less correlated to sales
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)

#plot the three with sales to see the relation

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_ylabel("Radio Spend")
axes[1].set_title("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_ylabel("Newspaper Spend")
axes[2].set_title("Sales")
plt.tight_layout();
```

```python
correlation = df.corr()
```

```python
sns.heatmap(correlation, vmax=1, square=True,annot=True, cmap='viridis')
#TV is more correlated (we can tell by the colors)
```
