```python
#Tips 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

#os.chdir('C:\\Users\Matthew...')
```

```python
tipsData = pd.read_csv('tips.csv')
```

```python
def gender(value):
        if value == "Female":
            return 1
        else:
            return 0
```

```python
tipsData["gender"] = tipsData["sex"].apply(gender)
```

```python
def weekend(value):
        if value == 'Mon':
            return 0
        elif value == 'Tues':
            return 0
        elif value == 'Wed':
            return 0
        elif value == 'Thur':
            return 0
        elif value == 'Fri':
            return 1
        elif value == 'Sat':
            return 1
        elif value == 'Sunday':
            return 1
        else:
            return -1
```

```python
tipsData['weekend'] = tipsData['day'].apply(weekend)
```
