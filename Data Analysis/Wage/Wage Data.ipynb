```python
WageDF['maritl'] = WageDF['maritl'].str.slice(0,1).astype(int) #replace columns with slices as num from front
WageDF['race'] = WageDF['race'].str.slice(0,1).astype(int)
WageDF['education'] = WageDF['education'].str.slice(0,1).astype(int)
WageDF['region'] = WageDF['region'].str.slice(0,1).astype(int)
WageDF['jobclass'] = WageDF['jobclass'].str.slice(0,1).astype(int)
WageDF['health'] = WageDF['health'].str.slice(0,1).astype(int)
WageDF['health_ins'] = WageDF['health_ins'].str.slice(0,1).astype(int)
```

```python
WageDF.describe()
```

```python
WageDF.tail(25)
```

```python
WageDF.to_csv(path + '\\' + 'Wage_New.csv', index=False, sep=',', header=True)#recreate the csv with new columns
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(15,8))
sns.scatterplot(x='age', y='wage', data=WageDF, ax=ax1) #show age and salary correlation
sns.scatterplot(x='year', y='wage', data=WageDF, ax=ax2) #show the year and wage correlation
sns.boxplot(x='education', y='wage', data=WageDF, ax=ax3)
ax2.set_ylabel("")
ax3.set_ylabel("")
ax3.set_xlabel("education")
```
